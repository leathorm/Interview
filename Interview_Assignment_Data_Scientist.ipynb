###**Q7: Was there a difference, on 28 July 2020, between the average salaries of the men and women who still work there?**

```python
# date to check
check_date = pd.to_datetime('2020-07-28')

# filter employees that worked on 28 July 2020
employees_working = df[
    (pd.to_datetime(df['Start_Month']) <= check_date) &
    (pd.to_datetime(df['Stop_Month']).fillna(pd.Timestamp.max) >= check_date)
]

# calculate average salary for men and women
average_salary_men = employees_working[employees_working['Gender'] == 'M']['Salary'].mean()
average_salary_women = employees_working[employees_working['Gender'] == 'F']['Salary'].mean()

# Show results
print(f"Average salary of men on 28 July 2020: {average_salary_men:.2f}")
print(f"Average salary of women on 28 July 2020: {average_salary_women:.2f}")

# Calculate difference
salary_difference = average_salary_men - average_salary_women
print(f"Difference in average salary between men and women: {salary_difference:.2f}")
```
